```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
```

```python
plt.figure()
plt.plot(x,y, "r--")
plt.xlabel("time x(s)")
plt.ylabel("y=sin(x)")
plt.title("simple plot of y=sin(x)", fontsize=20)
plt.show()
```

```python
from collections import namedtuple
Player = namedtuple("Players", ["name", "score"])
p = Player("Kobe", 81)
p
```

```python
from types import SimpleNamespace

ns = SimpleNamespace(name="Kobe", score=81)
print(ns.name) # ns["name"]会报错
ns.father = "Joe" # 相对于namedtuple，SimpleNamespace可以动态添加属性
print(ns.father)
```
