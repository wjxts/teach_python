```python
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns #seaborn是基于matplotlib的数据集分布可视化库
```

```python
# plt.plot(x, y) x/y一般是list或np.array
# 50kg
fontdict = {'family': 'Times New Roman', 'weight': 'bold', 'size': 16}
x = np.linspace(1.6, 1.9, 21)
y1 = 50 / x**2
y2 = 80 / x**2
standard_low = 18.5*np.ones_like(x)
stardard_high = 23.9*np.ones_like(x)
# ax = plt.figure(figsize=(6, 4))
plt.plot(x, y1, label="50kg", marker='o')
plt.plot(x, y2, label="80kg", marker='*')
plt.plot(x, standard_low, color='r', linestyle='--')
plt.plot(x, stardard_high, color='r', linestyle='--')
plt.xlabel("height (m)", fontdict=fontdict)
plt.ylabel("BMI (kg/m^2)", fontdict=fontdict)
plt.title("BMI vs height", fontdict=fontdict)
plt.xticks(fontproperties='Times New Roman', size=14, weight='bold')
plt.yticks(fontproperties='Times New Roman', size=14, weight='bold')
color = plt.cm.hot(0.2) #(255/255,0,0)
fontdict_text = {'family': 'Times New Roman', 'weight': 'bold', 'size': 16, 'color': color}
plt.text(x=1.6, y=24.5, s="upper line", fontdict=fontdict_text) #按照data坐标系给的
plt.text(x=1.8, y=19.0, s="lower line", fontdict=fontdict_text)
# plt.grid()
plt.legend(prop=fontdict)
# plt.savefig("bmi.pdf", dpi=300)
```

```python
plt.figure(figsize=(12, 4))
ax = plt.subplot(1, 2, 1)
plt.plot(x, y1, label="50kg", marker='o', 
         markerfacecolor='none', markeredgecolor='black', markeredgewidth=2, markersize=5)
# plt.grid(axis='both', linestyle='--')
# ax.spines[['left', 'bottom']].set_visible(False)
# plt.xticks([])
# plt.yticks([])
ax.spines[['top', 'right']].set_visible(False)
plt.legend(prop=fontdict)
plt.subplot(1, 2, 2)
plt.plot(x, y2, label="80kg", marker='*', color='y')
# 虚线grid 
plt.grid(axis='both', linestyle='--')
plt.tight_layout() # 自动调整子图间距
```

```python
# plt.scatter(x, y) x/y一般是list或np.array
plt.scatter(x, y1)
```

```python
# bar plot 
x = np.linspace(1, 5, 5)
y1 = np.random.randint(1, 20, 5)
y2 = np.random.randint(1, 20, 5)
plt.bar(x, y1, width=0.3, color='#ff7f0e', label='Men')
plt.bar(x+0.3, y2, width=0.3, color='#2ca02c', label='Women')
plt.legend(loc='upper right')
plt.ylim(0, 21)
plt.xticks(x+0.15, ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], rotation=45)
```

```python
# pie plot
x = np.random.randint(1, 10, 5)
plt.pie(x, labels=['A', 'B', 'C', 'D', 'E'], autopct='%1.1f%%', explode=[0, 0, 0, 0, 0])
```

```python
# histogram plot 
x = np.random.randn(10000)
plt.hist(x, bins=20, color='r', alpha=0.5, density=True) # alpha: 透明度
# distribution plot 
sns.distplot(x, bins=20, color='b', hist=True, kde=True, rug=False, fit=None, hist_kws=None, kde_kws=None, rug_kws=None, fit_kws=None, vertical=False, norm_hist=False, axlabel='x', label='x', ax=None)
```

```python
# 2D surface plot 
x = np.linspace(-3, 3, 101)
y = np.linspace(-3, 3, 101)
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2-Y**2)
plt.contourf(X, Y, Z, cmap='hot')
```

```python
x = np.linspace(-3, 3, 101)
y = np.linspace(-3, 3, 101)
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2-Y**2)
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
 
# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
