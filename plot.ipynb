```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#x = np.arange(10)
x = np.linspace(0, np.pi*4, 101)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure()
plt.plot(x, y1, "--", label="sin(t)")
plt.plot(x, y2, "*", label="sin(t)")
plt.xlabel("time (s)")
plt.ylabel("magnitute")
plt.title("magnitute vs t")
#plt.grid()
plt.legend()
```
